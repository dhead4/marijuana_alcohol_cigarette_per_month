# Part 1: Substance abuse rate in adolescence (aged 12-17) per state
## Analyzing your data with bar graphs

### Starting off, you want to gather your data and input it into an Excel file and save it as a .csv to import it into your jupyter notebook

```python
#Check to make your file you are trying
#to access is in the right location

from os import listdir
listdir()
```

```python
#Now that we've checked and made sure
#that our document is accessible,
#we can import the data into a chart
#here in our notebook with pandas

import pandas as pd
df = pd.read_csv('marijuna_alcohol_cigarette_consumption_month_12-17.csv')

#To display a snippet of your data
#you can use df.head()

df.head()
```

```python
#Everything is now squared away and we can
#now start the creation of our bar graph!
#First you'll want to import matplotlib
#which will give you access to creating countless
#charts for visualizing your data

import matplotlib.pyplot as plt

#Now that you can start making your
#bar graph, let's determine our axes

x = list(df['state'])
y = list(df['marijuana_past_month'])

#This next bit is optional, but you can
#also change the size of your display

f, ax = plt.subplots(1, 1, figsize = (15, 10))

#The next step is to add labels

#you can also change parts of
#your graph to your liking

plt.bar('state', 'marijuana_past_month', data = df, color = "Green")
plt.xlabel("State", fontsize = 20)
plt.xticks(rotation = 90, fontsize=12.5)
plt.ylabel("Prevalence Estimates", fontsize = 20)
plt.yticks(fontsize=12.5)

#Finally, we will give it it's title,
#save it, and display it!

plt.title("Fig. 1: Marijuana Usage in Past Month by State (ages 12-17)", fontsize = 25)
plt.savefig("marijuna_consumption_per_month_12-17.png", bbox_inches='tight', dpi=320)

plt.show()
```
