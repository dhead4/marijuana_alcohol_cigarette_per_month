```python
import pandas as pd
df = pd.read_csv('marijuna_alcohol_cigarette_consumption_month_12-17.csv')

df.head()
```

```python
import matplotlib.pyplot as plt

x = list(df['state'])
y = list(df['marijuana_past_month'])

f, ax = plt.subplots(1, 1, figsize = (15, 10))

plt.bar('state', 'marijuana_past_month', data = df, color = "Green")
plt.xlabel("State", fontsize = 20)
plt.xticks(rotation = 90, fontsize=12.5)
plt.ylabel("Prevalence Estimates", fontsize = 20)
plt.yticks(fontsize=12.5)
    
plt.title("Figure 1: Marijuana Usage in Past Month by State (ages 12-17)", fontsize = 25)

plt.savefig("marijuna_consumption_per_month_12-17.png", bbox_inches='tight', dpi=320)

plt.show()
```
